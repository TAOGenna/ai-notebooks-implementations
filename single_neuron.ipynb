# Single Neuron 
### Quadratic cost function

```python
# a single neuron trying to learn how to get number 0 from 1
# that is, always an input of 1 
# model from Michael Nielsen's book 

import numpy as np 
from tqdm import trange, tqdm
import matplotlib.pyplot as plt
from IPython.display import clear_output

input = 1

weight = 2
b = 2
learning_rate = 0.15

def sigmoid(x):
    return 1/(1+np.exp(-x))

def der_sigmoid(x):
    return sigmoid(x)*(1-sigmoid(x))

training_loop = 300

losses = []
accuracies = [] 

for i in (t := trange(training_loop)):
    # forward pass
    z = input*weight + b
    out = sigmoid(z)
    loss = (0-out)**2/2

    losses.append(loss)
    accuracies.append(out)
    # backward pass
    dCdw = out*der_sigmoid(z)
    dCdb = out*der_sigmoid(z)
    weight = weight - learning_rate*dCdw
    b = b - learning_rate*dCdb

print('prediction ', sigmoid(input*weight+b))

plt.plot(losses)
plt.plot(accuracies)
plt.show()
```
